## 4. 전처리 내용을 복습해 봅시다

제거한 값들을 복습해 봅시다

```
df2 = pd.read_csv('https://bit.ly/ds-house-price')
```

```
# 콤마가 있는 경우
df2.iloc[2125]
```

```
df2 = df2.rename(columns={"분양가격(㎡)" : '분양가격'})
df2
```

```
df2['분양가격'] = df2['분양가격'].str.replace(",","")
df2.iloc[2125]
```

```
# -가 있는 경우
df2.loc[df2['분양가격']=='-']
df2['분양가격'] = df2['분양가격'].str.replace("-","")
df2['분양가격'] = df2['분양가격'].fillna(0)
```

```
df2.loc[df2['분양가격']=='-']
```

```
# 공백이 2개 들어간 경우
df2.loc[df2['분양가격']=='  ']
df2['분양가격'] = df2['분양가격'].str.replace("  ","")
df2['분양가격'] = df2['분양가격'].fillna(0)
```

```
df2.loc[df2['분양가격']=='  ']
```

```
df2.loc[df2['분양가격']=='']
df2.loc[df2['분양가격']=='', '분양가격'] = 0
df2['분양가격'] = df2['분양가격'].fillna(0)
```

```
df2.loc[df2['분양가격']=='']
```

```
df2['분양가격'] = df2['분양가격'].astype(int)
```
