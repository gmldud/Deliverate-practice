```python
def get_inthewear_review(Total_Review):
    url_template = "https://subscription.inthewear.com/api/subscription/reviews?limit=10&offset={}"
    offset = 0
    reviews = []
    while True :
        url = url_template.format(Total_Review)
        inthewear_result = requests.get(url)
        data = inthewear_result.json()
        if not data :
            break
            
        reviews.extend(data)
        offset += 100
        
    return reviews
```

```python
import re
import requests
from bs4 import BeautifulSoup

BASE_URL = "https://movie.naver.com/movie"

# 리뷰들이 담긴 리뷰 리스트를 리턴해주는 함수
def get_reviews(movie_code, page_num=1):
    review_url = f"{BASE_URL}/point/af/list.naver?st=mcode&sword={movie_code}&target=after&page={page_num}"
    review_list = []

    page = requests.get(review_url)
    soup = BeautifulSoup(page.content, 'html.parser')

    review_body = soup.find_all('td', class_='title')

    for lst in review_body:
        review_list.append({'review_text' : lst.get_text().split('\n')[5],
                            'review_star' : lst.get_text().split('\n')[3].split('중')[1]})

    return review_list

# 총 스크레이핑할 리뷰 개수를 받아 해당 개수만큼 리뷰 항목이 담긴 리뷰 리스트를 리턴
def scrape_by_review_num(movie_title, review_num):
    review_lst = []

    movie_code = get_movie_code(movie_title)
    for i in range(int(review_num/10)+1):
        review_lst.extend(get_reviews(movie_code, i+1))

    reviews = review_lst[:review_num]

    return reviews

# 페이지 수를 기준으로 리뷰를 스크레이핑하는 함수
def scrape_by_page_num(movie_title, page_num=10):
    reviews = []

    movie_code = get_movie_code(movie_title)
    for i in range(page_num):
        reviews.extend(get_reviews(movie_code, i))

    return reviews
```
